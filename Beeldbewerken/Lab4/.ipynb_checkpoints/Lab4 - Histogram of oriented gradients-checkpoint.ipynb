## Histogram of oriented gradients

--Text--

```python
def subImage(f, yRange, xRange, yAnchor, xAnchor):
    
    return f[xAnchor:xAnchor+xRange:, yAnchor:yAnchor+yRange:]

bolt = resize(color.rgb2gray(subImage(imread('bolt2.png'), 100, 200, 205, 95)), (128, 64))

plt.imshow(bolt, cmap='gray')
plt.show()
```

```python
def cart2polar(fx, fy):
    
    mag = np.sqrt(fx * fx + fy * fy)
    angle = np.arctan(fy / fx)
    
    return mag, angle
    
def HOG8x8(fx, fy):
    
    mags, angles =  cart2polar(fx, fy)
    shapeX, shapeY = fx.shape
    mags = mags.reshape(shapeX * shapeY)
    angles = angles.reshape(shapeX * shapeY)
    # keep angles between 0 and pi
    angles[angles < 0] += np.pi
    angles[angles > np.pi] -= np.pi
    
    return np.histogram(angles, bins=np.linspace(0, np.pi, 10), weights=mags)
    

gx = np.ones((8,8))*np.sin(2 * (np.pi/3))
gy = np.ones((8,8))*np.cos(2 * (np.pi/3))
hist, bin_edges = HOG8x8(gx, gy)

binWidth = (bin_edges[0] + bin_edges[1])

plt.bar(bin_edges[:-1] + binWidth / 2, hist, width=binWidth)
plt.xlim(min(bin_edges) - 1, max(bin_edges) + 1)
plt.show()  
```

```python
from numpy.linalg import norm

def HOGblock(h1, h2, h3, h4):
    
    blockVector = np.hstack((h1, h2, h3, h4))
    return (blockVector / norm(blockVector))

t = [np.array([1, 2, 3, 4, 5, 6]) for _ in range(5)]

print(HOGblock(t[0], t[2], t[3], t[4]).shape)
```

```python
from skimage.filters import sobel_h, sobel_v

def rose(HOG, pos, ax):
    
    angles = np.linspace(0, np.pi, 10)
    
    for angle, length in zip(HOG, angles):
        ax.plot([pos[0] - length * np.sin(angle),pos[0] + length * np.sin(angle)],
                [pos[1] - length * np.cos(angle), pos[1] + length * np.cos(angle)])

def HOGimage(f):
    
    # Convert image to grayscale for use in sobel filter
    f = color.rgb2gray(f)
    
    # Calculate gradients of image
    fx = sobel_h(f)
    fy = sobel_v(f)
    
    # Pick out subimage
    bolt_fx = resize(subImage(fx, 100, 200, 205, 95), (128, 64))
    bolt_fy = resize(subImage(fy, 100, 200, 205, 95), (128, 64))
    
    # Set up iterator
    ys = [n for _ in range(9) for n in range(0, 65, 8)]
    xs = [n for n in range(0, 129, 8) for _ in range(9)]
    
    # Divide gradient image into blocks
    blocks_fx = [bolt_fx[x:x+8:,y:y+8:] for x, y in zip(xs, ys)]
    blocks_fy = [bolt_fy[x:x+8:,y:y+8:] for x, y in zip(xs, ys)]

    # Calculate HOG for each 8x8 cell
    HOGs = [HOG8x8(xs, ys)[0] for xs, ys in zip(blocks_fx, blocks_fy)]
    
    # Normalise HOGs over 16x16 blocks
    HOGblocks = [HOGblock(HOGs[x + y], HOGs[x + y + 1], HOGs[x + y + 7], HOGs[x + y + 8]) for x in range(8) for y in range(16)]

    positions = [(x, y) for x in np.linspace(4, 60, 8) for y in np.linspace(4, 124, 16)]
    fig, ax = plt.subplots(1, 1)
    ax.imshow(bolt_fx, cmap='gray')
    
    # Plot rose of directions for each HOG
    for HOG, pos in zip(HOGblocks, positions):
        rose(HOG, pos, ax)

    plt.show()
    
HOGimage(imread('bolt2.png'))
```
