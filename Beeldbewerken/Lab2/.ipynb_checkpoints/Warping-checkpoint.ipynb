```python
import numpy as np
import skimage as sk
import matplotlib.pyplot as plt

from scipy.ndimage import imread
from skimage.transform import warp

%matplotlib notebook
```

```python
def affineTransform(f, x1, y1, x2, y2, x3, y3, width, height):
    
    M = np.array([[x1, y1, 1, 0, 0, 0],
                  [0, 0, 0, x1, y1, 1],
                  [x2, y2, 1, 0, 0, 0],
                  [0, 0, 0, x2, y2, 1],
                  [x3, y3, 1, 0, 0, 0],
                  [0, 0, 0, x3, y3, 1]])
    
    q = np.array([0, 0, width - 1, 0, 0, height - 1])
    
    p = np.linalg.lstsq(M,q)
    p = p[0].reshape((2, 3))
    p = np.vstack((p, np.array([0, 0, 1])))
    
    p = np.linalg.inv(p)
    
    f_trans = warp(f, p)
    
    return f_trans

# Simple mouse click function to store coordinates
def onclick(event):
    global ix, iy
    ix, iy = event.xdata, event.ydata

    # print 'x = %d, y = %d'%(
    #     ix, iy)

    # assign global variable to access outside of function
    global coords
    coords.append((ix, iy))

    # Disconnect after 2 clicks
    if len(coords) == 2:
        fig.canvas.mpl_disconnect(cid)
        plt.close(1)
    return
    
    
f = imread('cameraman1.png')
f_trans = affineTransform(f, -100, 0, 256, 100, 100, 256, 256, 256)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.imshow(f)
coords = []
cid = fig.canvas.mpl_connect('button_press_event', onclick)


print(coords)
    
```
